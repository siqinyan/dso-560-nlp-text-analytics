```python
import pandas as pd
```

```python
data_orig = pd.read_csv('./Data/steam_reviews.csv')
data_orig
```

```python
data_orig = data_orig.iloc[:, 1:]
data_orig.describe()
```

```python
data_orig.shape
```

```python
data_orig.groupby(['recommended']).count()['app_id']
```

> filter reviews by language

```python
data = data_orig.loc[data_orig['language']=='english']
data.shape
```

```python
data.groupby(['recommended']).count()['app_id']
```

> in order to train efficiently on local machines, we decided to randomly select 6w records

```python
data_final = data.groupby(['recommended'], group_keys=False).apply(lambda x: x.sample(n=35000, 
                                                                                      random_state=41))
data_final
```

```python
data_final.groupby(['recommended']).count()['app_id']
```

```python
data_final.to_csv('./Data/steam_final.csv', index=False)
```
